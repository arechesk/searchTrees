```python
import pandas as pd
```

```python
df = pd.read_csv("./output.csv", sep=',')
print(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

## Insert 

```python

dsInsertTreap = df.loc[df['Algorithm'] == 'Treap insert']
dsInsertSplay = df.loc[df['Algorithm'] == 'Splay insert']
dsInsertMap = df.loc[df['Algorithm'] == 'Map insert']
dsInsertAVL = df.loc[df['Algorithm'] == 'AVL insert']
```

```python
First = ' count'
Second = ' time'
plt.rcParams['figure.figsize'] = [20, 10]
plt.plot(dsInsertTreap[First], dsInsertTreap[Second] / 100, label = 'Treap')
plt.plot(dsInsertSplay[First], dsInsertSplay[Second] / 100, label = 'Splay')
plt.plot(dsInsertMap[First], dsInsertMap[Second] / 100, label = 'Map')
plt.plot(dsInsertAVL[First], dsInsertAVL[Second] / 100, label = 'AVL')
plt.title('Time vs. Size of tree Insert')
plt.xlabel('Size of tree (elements)')
plt.ylabel('Time')
plt.legend()
# plt.savefig('Insert.png')
pass
```
